```c++
cout << "Enter a positive number: 25 " << endl;
int number = 25;
//cin >> number;
if (number < 0)
    cout << "Error! Number is not positive.";
else {
    long long factorial = 1;
    for (long long i = 1; i <= number; i++)
        factorial *= i;
    cout << "The factorial of " << number << ": " << factorial;
}
//Enter a positive number: 20
//The factorial of 20: 2432902008176640000
```

```c++
int temperatures[] = {60, 80, 90};
auto sum = 0;
for (auto temperature: temperatures)
    sum += temperature;
int count = sizeof(temperatures) / sizeof(temperatures[0]);
double average = (double) sum / count;
cout << average;
```
