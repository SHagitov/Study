```python
d = dict()
with open('26-96.txt') as f:
    N = int(f.readline())
    for _ in range(N):
        x, y = map(int, f.readline().split())
        d[y] = d.get(y, []) + [x // 10 if x >= 0 else -(abs(x) // 10)]

ans = [0, 0, 0]
ml = None
for x, y in sorted(d.items()):
    if len(y) >= ans[1]:
        ans = [x, len(y), len(set(y))]
print(ans[0], ans[2])
```
